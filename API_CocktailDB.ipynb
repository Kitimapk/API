```python
# import libralies
from requests import get
from time import sleep
import csv
```

```python
# example of get data from public API cocktails (https://www.thecocktaildb.com/api.php)

url = "http://www.thecocktaildb.com/api/json/v1/1/search.php?s=margarita"
res = get(url)
data = res.json()
data
```

```python
# create a list of popular cocktails
s = ["margarita", "negroni", "mojito", "daiquiri", "manhattan"]
for name in s:
  print(name)
  sleep(2)
```

```python
# generate url and run loop
for name in s:
  url = f"http://www.thecocktaildb.com/api/json/v1/1/search.php?s={name}"
  print(url)
  sleep(1)
```

```python
# creat API cocktail from a list value, run loop, and keep data in a list
s = ["margarita", "negroni", "mojito", "daiquiri", "manhattan"]

data = []

for name in s:
  url = f"http://www.thecocktaildb.com/api/json/v1/1/search.php?s={name}"
  res = get(url)
  print(res.json())
  res_js = res.json()
  name = res_js["drinks"][0]["strDrink"]
  category = res_js["drinks"][0]["strCategory"]
  tags = res_js["drinks"][0]["strTags"]
  instruction = res_js["drinks"][0]["strInstructions"]
  datail = [name, category, tags, instruction]
  data.append(datail)
  sleep(2)

print(data)
```

```python
# save data in csv file
header = ["name", "category", "tags", "instruction"]
with open("cocktails.csv", "w") as file:
  writer = csv.writer(file)
  writer.writerow(header)
  writer.writerows(data)
```
